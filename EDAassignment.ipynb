```python
df = pd.read_csv('BIKE DETAILS.csv')
df.head()
```

```python
# prompt: using datarame df:  1. What is the range of selling prices in the dataset?

# Find the minimum and maximum selling prices.
min_price = df['selling_price'].min()
max_price = df['selling_price'].max()

# Print the range.
print(f"The range of selling prices is from {min_price} to {max_price}.")
```

```python
# prompt: using datarame df: 2. What is the median selling price for bikes in the dataset?

median_price = df['selling_price'].median()
print(f"The median selling price for bikes is: {median_price}")
```

```python
# prompt: using datarame df: 3. What is the most common seller type?

seller_counts = df['seller_type'].value_counts()
most_common_seller = seller_counts.idxmax()
print(f"The most common seller type is: {most_common_seller}")


```

```python
# prompt: 4. How many bikes have driven more than 50,000 kilometers

# Assuming 'km_driven' column represents kilometers driven.
bikes_over_50k = df[df['km_driven'] > 50000]
num_bikes_over_50k = len(bikes_over_50k)

print(f"Number of bikes that have driven more than 50,000 kilometers: {num_bikes_over_50k}")
```

```python
# prompt: 5. What is the average km_driven value for each ownership type?

# Calculate the average km_driven for each ownership type
average_km_by_ownership = df.groupby('owner')['km_driven'].mean()

average_km_by_ownership
```

```python
# prompt: using datarame df: 6. What proportion of bikes are from the year 2015 or older?

older_than_2015_mask = df['year'] <= 2015
proportion_older_than_2015 = older_than_2015_mask.sum() / len(df)
print(f"Proportion of bikes from 2015 or older: {proportion_older_than_2015}")
```

```python
# prompt: using datarame df: 7. What is the trend of missing values across the dataset?

# Calculate the percentage of missing values for each column
missing_percentage = df.isnull().sum() * 100 / len(df)

# Create a bar plot to visualize the trend
plt.figure(figsize=(12, 6))
sns.barplot(x=missing_percentage.index, y=missing_percentage.values)
plt.xticks(rotation=90)
plt.ylabel('Percentage of Missing Values')
plt.title('Trend of Missing Values Across the Dataset')
plt.show()

# Print the missing value counts and percentages
print(df.isnull().sum())
missing_percentage
```

```python
 # prompt: using datarame df: 8. What is the highest ex_showroom_price recorded, and for which bike?

# Find the highest ex_showroom_price and the corresponding bike name.
highest_ex_showroom_price = df['ex_showroom_price'].max()
bike_with_highest_price = df.loc[df['ex_showroom_price'] == highest_ex_showroom_price, 'name'].iloc[0]

print(f"The highest ex_showroom_price is {highest_ex_showroom_price} for the {bike_with_highest_price} bike.")
```

```python
# prompt: 9. What is the total number of bikes listed by each seller type

seller_bike_counts = df.groupby('seller_type')['name'].count()
seller_bike_counts
```

```python
# prompt: 10. What is the relationship between selling_price and km_driven for first-owner bikes?

# Filter for first-owner bikes
first_owner_bikes = df[df['owner'] == 'First Owner']

# Create a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='km_driven', y='selling_price', data=first_owner_bikes)
plt.title('Relationship between Selling Price and Kilometers Driven for First-Owner Bikes')
plt.xlabel('Kilometers Driven')
plt.ylabel('Selling Price')
plt.show()

# Calculate the correlation coefficient
correlation = first_owner_bikes['km_driven'].corr(first_owner_bikes['selling_price'])
print(f"The correlation coefficient between km_driven and selling_price for first-owner bikes is: {correlation}")
```

```python
# prompt: 11. Identify and remove outliers in the km_driven column using the IQR method

# Calculate Q1, Q3, and IQR for 'km_driven'
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['km_driven'] < lower_bound) | (df['km_driven'] > upper_bound)]

# Remove outliers
df_no_outliers = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

# Print the number of outliers removed
print(f"Number of outliers removed: {len(outliers)}")
```

```python
# prompt: 12. Perform a bivariate analysis to visualize the relationship between year and selling_price.

import matplotlib.pyplot as plt
import seaborn as sns

# Create the bivariate plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='year', y='selling_price', data=df)
plt.title('Relationship between Year and Selling Price')
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.show()
```

```python
# prompt: 13. What is the average depreciation in selling price based on the bike's age (current year-manufacturing year)?

# Calculate the bike's age
df['bike_age'] = pd.to_datetime('today').year - df['year']

# Group by bike age and calculate the average selling price for each age group
average_price_by_age = df.groupby('bike_age')['selling_price'].mean()

# Calculate the depreciation for each age group
depreciation_by_age = average_price_by_age.diff()

# Calculate the average depreciation
average_depreciation = depreciation_by_age.mean()

print(f"The average depreciation in selling price based on the bike's age is: {average_depreciation}")
```

```python
# prompt: 14. Which bike names are priced significantly above the average price for their manufacturing year?

# Calculate the average selling price for each year
average_price_by_year = df.groupby('year')['selling_price'].mean()

# Identify bikes priced significantly above the average for their year
significantly_above_average = df[df['selling_price'] > df['year'].map(average_price_by_year) * 1.5]

# Print the names of these bikes
print(significantly_above_average['name'])
```

```python
# prompt: 15. Develop a correlation matrix for numeric columns and visualize it using a heatmap.

# Assuming 'df' is your DataFrame from the previous code.

# Select only numeric columns for correlation analysis
numeric_cols = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation_matrix = numeric_cols.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Columns')
plt.show()
```
