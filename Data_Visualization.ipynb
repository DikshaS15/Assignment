####5.Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate correlated data
np.random.seed(42)
n_samples = 100
cov_matrix = [[1, 0.8, 0.5], [0.8, 1, 0.7], [0.5, 0.7, 1]]
data = np.random.multivariate_normal(mean=[0, 0, 0], cov=cov_matrix, size=n_samples)

# Create a DataFrame
df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3'])

# Calculate the correlation matrix
corr_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# 2.Heatmap for Sales Data
# Heres the code to generate a heatmap visualizing the variation in sales across different months and days:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the random seed for reproducibility
np.random.seed(20)

# Generate the dataset for sales data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data)

# Pivot the data to create a matrix format for the heatmap
heatmap_data = df_sales.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap='viridis')
plt.title('Variation in Sales Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set random seed for reproducibility
np.random.seed(25)

# Create the DataFrame
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create the bubble chart using Matplotlib
plt.figure(figsize=(10, 6))

# Plot the bubbles with size proportional to population
plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.7)

# Add labels for each bubble
for i in range(len(df)):
    plt.annotate(df['Country'][i], (df['GDP'][i], df['Population'][i]))

# Set labels and title
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Country Comparison: GDP vs. Population')

# Show the plot
plt.grid(True)
plt.show()
```

####1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Display the plot in the notebook
output_notebook()
show(p)
```

####2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
sizes = np.random.rand(100) * 50 + 10  # Sizes between 10 and 60
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], 100)

# Create a ColumnDataSource
source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a new plot
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add scatter renderer
p.scatter('x', 'y', size='sizes', color='colors', fill_alpha=0.6, source=source)

# Display the plot in the notebook
output_notebook()
show(p)
```

####3.Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import output_file
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Output to notebook
output_notebook()

# Create a new plot
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Count')

# Add bars
p.vbar(x=fruits, top=counts, width=0.9, color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

# Styling
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import transform
from bokeh.io import output_file
from bokeh.layouts import column

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare the data for Bokeh
data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
}

# Create a color mapper
color_mapper = LinearColorMapper(palette='Viridis256', low=data['value'].min(), high=data['value'].max())

# Create the figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', tools='')

# Add the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=data,
       fill_color=transform('value', color_mapper), line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0, 0))
p.add_layout(color_bar, 'right')

# Output to notebook and show
output_notebook()
show(p)
```
