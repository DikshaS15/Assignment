1. What is the difference between a function and a method in Python?

Ans-1. In Python, the terms "function" and "method" are often used interchangeably, but there's a subtle distinction:

### FUNCTION :

* A self-contained block of code that performs a specific task.
* Can be defined independently of any class or object.
* Can be called directly using its name.

### METHOD :

* A function that is defined within a class.
* Belongs to a specific object instance.
* Can be called using the dot notation on an object instance (e.g., object.method()).

### Key differences :

 1. SCOPE: Functions have global or local scope, while methods have class scope and can access the class's attributes and other methods.

2. BINDING: Functions are not bound to specific objects, while methods are bound to the object they are associated with.

3. INVOCATION: Functions are called directly, while methods are called through an object instance.









```python
# Example:

# Function
def greet(name):
    print("Hello, " + name + "!")

# Method
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print("Hello, my name is " + self.name + "!")

# Calling the function
greet("Alice")

# Creating an object of the Person class
person = Person("Bob")

# Calling the method on the object
person.greet()


```

In this example, greet is a function, while Person.greet is a method. The function can be called directly with any name, while the method can only be called on an instance of the Person class.

In summary, functions are standalone blocks of code, while methods are functions that are part of a class and are associated with specific objects.

### 2. Explain the concept of function arguments and parameters in Python.

Ans-2. In Python, function arguments and parameters are essential components for defining and using functions effectively. Here's an explanation of their concepts:

1. FUNCTION ARGUMENTS :

* **Definition**: Arguments are the actual values passed to a function when it is called. They are the concrete data that the function will work with.

* Values passed to a function when it is called.
* Can be positional arguments (specified by their order) or keyword arguments (specified by their names).
* Can be required (must be provided) or optional (have default values).
* Can be mutable (can be modified inside the function) or immutable (cannot be modified inside the function).

* **PURPOSE**: Arguments provide the specific information that a function needs to execute its logic. They fill in the placeholders defined by the parameters.

2. PARAMETERS :

* **Definition**: Parameters are the variables listed inside the parentheses in the function definition. They act as placeholders for the values that will be passed to the function when it is called.

* Can be required or optional.
* Can be mutable or immutable.

* **Purpose**: Parameters allow functions to be flexible and reusable. They define the data that a function needs to operate on, enabling the function to perform its intended task with different input values.


```python
# Example:

def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

# Calling the function with positional arguments:
greet("Aru")  # Output: Hello, Aru!

# Calling the function with keyword arguments:
greet(name="Riya", greeting="Hi")  # Output: Hi, Riya!

# Calling the function with an optional argument:
greet("Charlie")  # Output: Hello, Charlie!
```

In this example:

*  name and greeting are function parameters.
* "Aru", "Riya", "Charlie" are function arguments.
* greeting is an optional parameter with a default value of "Hello".

## Key Points:

* **Argument passing**: The values of arguments are passed to the corresponding parameters when the function is called.

* **Scope:** Parameters are local to the function, meaning they exist only within the function's body.

* **Mutable vs. immutable**: If a mutable argument is modified inside the function, the changes will be reflected outside the function as well. However, if an immutable argument is modified, the original value remains unchanged.

### Summary :

* Function arguments and parameters are essential components of Python functions. Arguments provide the data for the function to work with, while parameters define the variables that will receive those values. Understanding these concepts is crucial for writing effective and reusable Python code.

### 5.  What are iterators in Python and how do they differ from iterables?

Ans-5. **Iterators and Iterables in Python**

In Python, iterators and iterables are fundamental concepts for working with sequences of elements. While they are closely related, they have distinct roles and characteristics.

1. **ITERABLES**:

* **Definition**: An iterable is any object that can be iterated over, meaning its elements can be accessed one by one.

* **Examples**: Lists, tuples, strings, dictionaries, sets, and custom-defined objects that implement the __iter__ method.

* **Behavior**: When an iterable is passed to the iter() function, it returns an iterator.

2. **ITERATORS:**

* **Definition**: An iterator is an object that represents a sequence of values and can be used to iterate over that sequence.

* **Characteristics:**
* Implements the **__iter__** and **__next__ **methods.

* The __iter__ method returns the iterator itself.

* The __next__ method returns the next value in the sequence. When there are no more values, it raises a StopIteration exception.

#### Usage:
* To obtain an iterator from an iterable, use the **iter()** function.

* To retrieve the next value from an iterator, use the next() function.

* Iterators can be used in **for** loops, which automatically call **next()** until a **StopIteration** exception is raised.

Key Differences:

Purpose: Iterables are the source of elements, while iterators are the objects used to access those elements one by one.
Methods: Iterables typically implement the __iter__ method, while iterators implement both __iter__ and __next__.
Behavior: Iterables return iterators when passed to iter(), while iterators are used directly for iteration.



```python
# Example:

my_list = [1, 2, 3]  # Iterable

# Get an iterator from the list
my_iterator = iter(my_list)

# Iterate using the iterator
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3

# Trying to get the next element after the end raises StopIteration
try:
    print(next(my_iterator))
except StopIteration:
    print("Iteration is finished")

# my_list = [1, 2, 3]  # Iterable

# Get an iterator from the list
my_iterator = iter(my_list)

# Iterate using the iterator
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3

# Trying to get the next element after the end raises StopIteration
try:
    print(next(my_iterator))
except StopIteration:
    print("Iteration is finished")


# my_list = [1, 2, 3]  # Iterable

# Get an iterator from the list
my_iterator = iter(my_list)

# Iterate using the iterator
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3

# Trying to get the next element after the end raises StopIteration
try:
    print(next(my_iterator))
except StopIteration:
    print("Iteration is finished")
```

In summary, iterables provide the data to be iterated over, while iterators act as pointers that traverse the sequence, allowing you to access elements one at a time.

### 6.  Explain the concept of generators in Python and how they are defined.

Ans-6. ** GENERATORS**

In Python, generators are a special type of function that returns an iterator. Unlike regular functions that return a single value at a time, generators yield values one by one using the **yield** keyword. This allows for efficient memory usage, especially when dealing with large datasets.

* **DEFINITION**
* Generators are defined using the **def** keyword, just like regular functions. However, they contain at least one **yield** statement. When a generator function is called, it returns a generator object. This object is an iterator, which means it can be used to iterate over its elements using a **for** loop or the **next()** function.

```python
# Example:

def square_generator(n):
    for i in range(n):
        yield i ** 2

# Create a generator object
squares = square_generator(5)

# Iterate over the generator
for square in squares:
    print(square)
```

**Key points about generators:**

* **Efficiency**: Generators are more memory-efficient than regular functions that return lists or tuples, especially when dealing with large datasets.

* **Lazy evaluation**: Generators yield values one by one as needed, rather than generating all values upfront.

* **Iterators**: Generator objects are iterators, which means they can be used in for loops and with the next() function.

* **State preservation**: Generators maintain their state between calls, allowing them to resume execution from where they left off.

#### Use cases:
* **Data processing**: Generators are often used to process large datasets efficiently, as they avoid loading the entire dataset into memory at once.

* **Infinite sequences**: Generators can be used to create infinite sequences, such as Fibonacci numbers or prime numbers.

* **Custom iterators**: Generators can be used to create custom iterators for various data structures or algorithms.

#### Additional notes:

* Generators can be used with yield from to delegate iteration to other generators or iterables.

* Generators can be combined with yield from to create more complex iteration patterns.

* Generators can be used to implement coroutines, which are a more advanced form of concurrency in Python.

### 7. What are the advantages of using generators over regular functions?

Ans-7. **Advantages of Generators Over Regular Functions**

Generators are a special type of function in Python that return an iterator. They provide a powerful and efficient way to generate sequences of values on-the-fly, especially when dealing with large datasets or infinite sequences.

##### Here are some of the key advantages of using generators over regular functions:

1. MEMORY EFFICIENCY:

* **Lazy Evaluation**: Generators don't create the entire sequence of values upfront. Instead, they produce values one at a time as needed, leading to significant memory savings, especially when dealing with large datasets.

* **Avoidance of Unnecessary Calculations**: Generators can avoid unnecessary calculations if the entire sequence is not required. This can improve performance and reduce resource usage.

2. EFFICIENT ITERATION:

* **Built-in Iteration**: Generators are designed for iteration. They can be used directly in loops, making it easy to consume the generated values.

* **Optimized for Iteration**: Python's iteration protocol is optimized for generators, resulting in efficient traversal of the generated sequence.

3. INFINITE SEQUENCES:

* **Generation of Infinite Sequences**: Generators can be used to create infinite sequences, such as generating Fibonacci numbers or counting up to infinity. This is not possible with regular functions.

4. PIPELINE CREATION:

* **Data Pipelines**: Generators can be combined to create data pipelines, where the output of one generator is used as the input for another. This allows for efficient and modular data processing.

5. CONCISE AND READABLE CODE:

* **Clearer Intent**: Generators often result in more concise and readable code, especially when dealing with complex sequence generation logic.

```python
# Example:

def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

for num in fibonacci():
    if num > 1000:
        break
    print(num)
```

**In this example**, the **fibonacci** generator function produces an infinite sequence of Fibonacci numbers. The **yield** keyword pauses the function's execution and returns the current value, allowing the loop to iterate through the sequence.

#### Conclusion:
Generators offer several advantages over regular functions, including memory efficiency, laziness, simplicity, and the ability to handle infinite sequences. They are particularly useful when dealing with large datasets, complex computations, or iterative processes. By understanding and leveraging these advantages, you can write more efficient, readable, and maintainable Python code.

### 9.  Explain the purpose and usage of the `map()` function in Python.

Ans-9. The map() function in Python is a built-in function that applies a given function to each item of an iterable (like a list, tuple, or dictionary) and returns an iterator containing the results.

#### PURPOSE:

* To apply a function to each element of an iterable efficiently.

* To transform elements of an iterable into a new form.
To create new iterables based on existing ones.

#### USAGE:

* map(function, iterable)

* **function:** The function to be applied to each element.

* **iterable:** The iterable whose elements will be mapped.

```python
# Example:

numbers = [1, 2, 3, 4, 5]

# Square each number using map()
squared_numbers = map(lambda x: x**2, numbers)

# Convert the iterator to a list for printing
squared_numbers_list = list(squared_numbers)
print(squared_numbers_list)  # Output: [1, 4, 9, 16, 25]
```

**In this example:**

1. The lambda function is used to square each number.

2. The map() function applies the lambda function to each element of the numbers list.

3. The resulting iterator is converted to a list for printing.

#### Key Points:

* The map() function returns an iterator, not a list. You need to convert it to a list or other data structure if you want to work with its elements.

* The map() function is often used with lambda functions for concise and efficient operations.

* You can use multiple iterables as arguments to map(), but the function must take the same number of arguments as the number of iterables.

#### Additional notes:

* The map() function is a generator function, so it creates elements on-the-fly as needed. This can be more memory-efficient than creating a complete list upfront.

* If you need to apply a function to elements of multiple iterables simultaneously, consider using the zip() function in combination with map().

* By understanding the purpose and usage of the map() function, you can effectively apply it to various programming tasks in Python.

### 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans-10. the differences between the map(), reduce(), and filter() functions in Python:

1. map():

* **Purpose:** Applies a function to every element in an iterable (like a list, tuple, or set) and returns a new iterable containing the results.

* **Syntax:** map(function, iterable)

* **Behavior:** Takes a function and an iterable as input.
Applies the function to each element of the iterable.
Returns a new iterable containing the results of the function applied to each element.

```python
# Example:

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x**2, numbers)
print(list(squared_numbers))
```

2. reduce():

* **Purpose:** Applies a function to an iterable to reduce it to a single value.

* **Syntax:** reduce(function, iterable, initial_value=None)

* **Behavior:**

* Takes a function, an iterable, and optionally an initial value.

* Applies the function to the first two elements of the iterable.

* Applies the function to the result of the previous application and the next element of the iterable.

* Continues until all elements have been processed.

* Returns the final result.

```python
# Example:

from functools import reduce

numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

3. filter():

* **Purpose: ** Creates a new iterable containing only the elements from the original iterable that satisfy a given condition.

* **Syntax**: filter(function, iterable)

* **Behavior**:

* Takes a function and an iterable as input.

* Applies the function to each element of the iterable.

* Keeps only the elements for which the function returns True.

* Returns a new iterable containing the filtered elements.

```python
# Example:

numbers = [1, 2, 3, 4, 5]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))
```

**Key Differences:**

1. **map()**: Applies a function to each element and returns a new iterable with the results.

2. **reduce()**: Reduces an iterable to a single value using a function.

3. **filter()**: Filters elements based on a condition and returns a new iterable with the filtered elements.

#### **Choosing the Right Function:**

* Use map() when you want to apply a transformation to each element of an iterable.

* Use reduce() when you want to combine the elements of an iterable into a single value.

* Use filter() when you want to select elements from an iterable based on a condition.
